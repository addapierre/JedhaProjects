```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
pd.options.display.max_columns = None
```

```python
dating = pd.read_csv('Speed Dating Data.csv', encoding = 'ISO-8859-1')
```

```python
usecol = ['gender', 'match', 'age', 'age_o', 'income']
dating = dating.loc[:,usecol].copy()
```

```python
dating['age_diff'] = dating.age - dating.age_o
```

```python
dating_F = dating.loc[(dating.gender == 0), :]
dating_M = dating.loc[(dating.gender == 1), :]

dating_nomatch = dating.loc[(dating.match == 0), :]
dating_match = dating.loc[(dating.match == 1), :]


dating_F_match = dating_F.loc[dating_F.match == 1,:]
dating_F_nomatch = dating_F.loc[dating_F.match == 0,:]

dating_M_match = dating_M.loc[dating_M.match == 1,:]
dating_M_nomatch = dating_M.loc[dating_M.match == 0,:]

```
