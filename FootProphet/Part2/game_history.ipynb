```python
import pandas as pd
import numpy as np
from tqdm import tqdm
tqdm.pandas()

pd.options.display.max_columns=None
```

```python
clubs = pd.read_csv('data/clubs.csv')

games_21_22 = pd.read_csv('data/raw/DCU_games_21_22.csv')
games_20_21 = pd.read_csv('data/raw/DCU_games_20_21.csv')
games_19_20 = pd.read_csv('data/raw/DCU_games_19_20.csv')
```

```python
print(games_21_22.HomeTeam.sort_values().unique())
print(games_20_21.HomeTeam.sort_values().unique())
print(games_19_20.HomeTeam.sort_values().unique())

```

```python
games_21_22['season'] = '21/22'
games_20_21['season'] = '20/21'
games_19_20['season'] = '19/20'
```

```python
games = pd.concat([games_21_22,games_20_21,games_19_20]).reset_index(drop=True)
games.Date = pd.to_datetime(games.Date)
games = games.sort_values('Date').reset_index(drop = True)
```

```python
games
```

```python
games['HomeTeam_id'] = games.HomeTeam.apply(lambda x: clubs.loc[clubs.data_co_uk_club==x,'club_id'].values[0] if x in clubs.data_co_uk_club.values else None)
games['AwayTeam_id'] = games.AwayTeam.apply(lambda x: clubs.loc[clubs.data_co_uk_club==x,'club_id'].values[0] if x in clubs.data_co_uk_club.values else None)
```

```python
games
```

```python
games.to_csv('data/DCU_games.csv', index = False)
```
