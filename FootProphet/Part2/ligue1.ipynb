```python
#ordre des stats: GK, ATK, DEF, MID:
game_composition_ADM.poste.unique()
```

```python
ATK = game_composition_ADM.loc[game_composition_ADM.poste == 'Att.',:].sort_values('club_id', ascending=True)
DEF = game_composition_ADM.loc[game_composition_ADM.poste == 'Déf.',:].sort_values('club_id', ascending=True)
MID = game_composition_ADM.loc[game_composition_ADM.poste == 'Mil.',:].sort_values('club_id', ascending=True)
```

```python
clubs.loc[~clubs.club_id.isin(ATK.club_id),'club_id'].values[0]
```

```python
try:
    clubs.loc[~clubs.club_id.isin(ATK.club_id),'club_id'].values[0]
    for missing_club in clubs.loc[~clubs.club_id.isin(ATK.club_id),'club_id'].values:
        new_line = {'club_id':[missing_club],'poste':['Att.']}
        new_line.update({x:[0] for x in ATK.columns[2:]})
        new_line = pd.DataFrame(new_line)
        ATK = pd.concat([new_line,ATK]).sort_values('club_id',ascending=True)
except:
    print('no games with no attackers')
try:
    clubs.loc[~clubs.club_id.isin(DEF.club_id),'club_id'].values[0]
    for missing_club in clubs.loc[~clubs.club_id.isin(DEF.club_id),'club_id'].values:
        new_line = {'club_id':[missing_club],'poste':['Déf.']}
        new_line.update({x:[0] for x in DEF.columns[2:]})
        new_line = pd.DataFrame(new_line)
        DEF = pd.concat([new_line,DEF]).sort_values('club_id',ascending=True)
except:
    print('no games with no defensers')
try:
    clubs.loc[~clubs.club_id.isin(MID.club_id),'club_id'].values[0]
    for missing_club in clubs.loc[~clubs.club_id.isin(MID.club_id),'club_id'].values:
        new_line = {'club_id':[missing_club],'poste':['Mil.']}
        new_line.update({x:[0] for x in MID.columns[2:]})
        new_line = pd.DataFrame(new_line)
        MID = pd.concat([new_line,MID]).sort_values('club_id',ascending=True)
except:
    print('no games with no midfielders')
```
